#### Explore the distribution of categorical variables

```python

# Set a larger plot size for better visualization
plt.figure(figsize=(12, 8))

# Explore the distribution of "Country Code" with a bar chart
plt.subplot(3, 1, 1)
sns.countplot(x='Country Code', data=df, order=df['Country Code'].value_counts().index)
plt.title("Country Code Distribution")


# Set a larger plot size for better visualization
plt.figure(figsize=(10, 6))

# Explore the distribution of "City" with a pie chart
city_distribution = df['City'].value_counts()
top_cities = city_distribution.nlargest(10)  # Choose the top N cities for the pie chart

# Plot a pie chart for the top cities
plt.pie(top_cities, labels=top_cities.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 10 Cities Distribution")

# Show the plot
plt.show()

# Explore the distribution of "Cuisines" with a bar chart (top 10 cuisines for better readability)
plt.subplot(3, 1, 3)
top_cuisines = df['Cuisines'].value_counts().nlargest(10)
sns.barplot(x=top_cuisines.values, y=top_cuisines.index)
plt.title("Top 10 Cuisines Distribution")

# Adjust layout for better readability
plt.tight_layout()

# Show the plots
plt.show()
```

#### Identify the top cuisines and cities with the highest number of restaurants.

```python
# Set up the matplotlib figure
plt.figure(figsize=(14, 6))

# Plot Top Cuisines
plt.subplot(1, 2, 1)
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='viridis', orient='h')
plt.title('Top 10 Cuisines')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')

# Plot Top Cities
plt.subplot(1, 2, 2)
sns.barplot(x=top_cities.values, y=top_cities.index, palette='magma', orient='h')
plt.title('Top 10 Cities with the Highest Number of Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

#### Determine Correlation between Location and Rating:

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with latitude, longitude, and aggregate rating
scatter = ax.scatter(df['Latitude'], df['Longitude'], df['Aggregate rating'], c=df['Aggregate rating'], cmap='viridis', alpha=0.5)

# Add colorbar
cbar = fig.colorbar(scatter, ax=ax, label='Aggregate Rating')

# Set labels and title
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_zlabel('Aggregate Rating')
ax.set_title('Correlation between Location and Rating')

plt.show()
```
