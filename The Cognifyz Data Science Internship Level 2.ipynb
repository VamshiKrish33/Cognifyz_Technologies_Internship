### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the data set

```python
df=pd.read_csv('Dataset .csv')
```

```python
df.info()
```

```python
df
```

```python
total_restaurants = len(df)

table_booking_percentage = (df['Has Table booking'] == 'Yes').sum() / total_restaurants * 100
online_delivery_percentage = (df['Has Online delivery'] == 'Yes').sum() / total_restaurants * 100

print(f"Percentage of restaurants offering Table Booking: {table_booking_percentage:.2f}%")
print(f"Percentage of restaurants offering Online Delivery: {online_delivery_percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt

colors = ['pink', '#FCF6F5']
table_booking_percentage = 12.12
online_delivery_percentage = 25.66

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# Bar chart for Table Booking
axes[0].barh(['Table Booking', 'No Table Booking'], [table_booking_percentage, 100 - table_booking_percentage],
             color=colors)
axes[0].set_title('Percentage of Restaurants Offering Table Booking')
axes[0].set_xlabel('Percentage')

# Bar chart for Online Delivery
axes[1].barh(['Online Delivery', 'No Online Delivery'], [online_delivery_percentage, 100 - online_delivery_percentage],
             color=colors)
axes[1].set_title('Percentage of Restaurants Offering Online Delivery')
axes[1].set_xlabel('Percentage')

plt.tight_layout()
plt.show()
```

```python
avg_rating_with_table = df[df['Has Table booking'] == 'Yes']['Aggregate rating'].mean()
avg_rating_without_table = df[df['Has Table booking'] == 'No']['Aggregate rating'].mean()

print(f"Average rating with Table Booking: {avg_rating_with_table:.2f}")
print(f"Average rating without Table Booking: {avg_rating_without_table:.2f}")

average_ratings = [avg_rating_with_table, avg_rating_without_table]
labels = ['With Table Booking', 'Without Table Booking']

plt.pie(average_ratings, labels=labels, autopct='%1.1f%%', startangle=90,colors=colors,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'antialiased': True})

plt.title('Average Ratings with and without Table Booking')
plt.show()
```

```python
#Availability of online delivery with different price ranges

price_ranges = df['Average Cost for two'].apply(lambda x: 'Low' if x < 500 else 'Medium' if 500 <= x <= 1000 else 'High')
online_delivery_by_price_range = df.groupby(price_ranges)['Has Online delivery'].value_counts(normalize=True).unstack()

print("Online Delivery Availability by Price Range:")
print(online_delivery_by_price_range)


online_delivery_by_price_range.plot(kind='barh', stacked=True, color=['pink', 'white'], edgecolor='black')
plt.xlabel('Number of Restaurants')
plt.ylabel('Price Range')
plt.title('Online Delivery by Price Range')
plt.legend(['With Online Delivery', 'Without Online Delivery'])
plt.show()
```

```python
# Find the most common price range among all restaurants

most_common_price_range = df['Price range'].mode()[0]
print(f"The most common price range among all restaurants is: {most_common_price_range}")
```

```python
# Calculate the average rating for each price range

avg_rating_by_price_range = df.groupby('Price range')['Aggregate rating'].mean()

print("Average rating for each price range:")
print(round(avg_rating_by_price_range,3))
```

```python
avg_rating_by_price_range = df.groupby('Price range')['Aggregate rating'].mean()
highest_avg_rating_range = avg_rating_by_price_range.idxmax()

# Create a pie chart with the highest average rating highlighted
colors = ['pink' if range == highest_avg_rating_range else 'white' for range in avg_rating_by_price_range.index]

plt.pie(avg_rating_by_price_range, labels=avg_rating_by_price_range.index, autopct='%1.1f%%', startangle=90, colors=colors,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'antialiased': True})

plt.title('Average Rating by Price Range')
plt.show()

print(f"Highest Average Rating Price Range: {highest_avg_rating_range}")
```

```python
# Extracting length of restaurant name and address
df['Restaurant Name length'] = df['Restaurant Name'].apply(len)
df['Address length'] = df['Address'].apply(len)
```

```python
# Encoding categorical variables like 'Has Table Booking' and 'Has Online Delivery'
df['has table booking'] = df['Has Table booking'].replace({'Yes': 1, 'No': 0}).astype(int)
df['has online delivery'] = df['Has Online delivery'].replace({'Yes': 1, 'No': 0}).astype(int)
```

```python
# Displaying the modified dataset with new features
print("Modified Dataset with New Features:")
df # Show the first few rows of the modified dataset
```

```python
# Summary statistics of the new features
print("\nSummary Statistics of New Features:")
print(df[['Restaurant Name length', 'Address length', 'has table booking', 'has online delivery']].describe())
```
